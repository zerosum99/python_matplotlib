```python
a = np.arange(0,5,0.01)
```

```python
a.shape
```

## 그래프를 넣기

```python
fig = plt.figure(figsize=(10,12))
plt.subplot(411)
plt.plot(np.sqrt(a))

plt.subplot(423)
plt.plot(np.sin(a))

plt.subplot(424)
plt.plot(np.cos(a))

plt.subplot(425)
plt.plot(a**2)

plt.subplot(426)
plt.plot(np.exp(a))

plt.subplot(414)
plt.plot(np.tan(a))

plt.show()
```
